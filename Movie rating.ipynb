Top 10 movies

```python
top_movie = movie_data.loc[movie_data['Rating'].sort_values(ascending = False)[:10].index]
top_movie
```

```python
# Convert 'Duration' to string first, then remove ' min' and strip whitespace
movie_data.loc['Duration'] = movie_data['Duration'].astype(str).str.replace(' min', '').str.strip()

# Convert the 'Duration' column to numeric, forcing invalid entries to NaN
movie_data.loc[:, 'Duration'] = pd.to_numeric(movie_data['Duration'], errors='coerce')

# Filter the movies with 'Duration' of 180 minutes or more
long_movies = movie_data[movie_data['Duration'] >= 180]

# Display the 'Title' of the filtered movies
print(long_movies['Name'])
```

Highest avg Voting

```python
movie_data.columns
```

```python
movie_data.loc[:, 'Votes'] = pd.to_numeric(movie_data['Votes'], errors='coerce')
movie_data.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Votes',data=movie_data)
plt.title("Votes By Year")
```

```python
genre_counts = movie_data['Genre'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']

# Select the top N genres (e.g., top 5)
top_n_genres = genre_counts.head(5)
top_n_genres
```

```python
plt.figure(figsize=(4, 4))
plt.pie(top_n_genres['Count'], labels=top_n_genres['Genre'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Movie Genre Distribution', fontsize=16)
plt.axis('equal')
plt.show()
     
```
